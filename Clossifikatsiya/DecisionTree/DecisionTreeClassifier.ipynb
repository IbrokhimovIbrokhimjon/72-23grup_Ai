```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
# Iris datasetni yuklaymiz
iris = load_iris()
X = iris.data
y = iris.target

# Ma'lumotlarni train/test ga bo'lamiz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Modelni yaratamiz
model = DecisionTreeClassifier()

# Modelni o'qitamiz
model.fit(X_train, y_train)
```

```python
# Bashorat qilish
y_pred = model.predict(X_test)

# Aniqlikni baholaymiz
print("Aniqlik (accuracy):", accuracy_score(y_test, y_pred))
print("To‘liq hisobot:")

print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree Diagrammasi")
plt.show()
```
