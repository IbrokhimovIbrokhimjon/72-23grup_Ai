```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Train/test bo‘lish (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Model yaratish (100 ta daraxt bilan)
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Modelni o‘rgatish
model.fit(X_train, y_train)
```

```python
# Test to‘plamiga nisbatan bashorat
y_pred = model.predict(X_test)

# Natijalarni baholash
print("Aniqlik (Accuracy):", accuracy_score(y_test, y_pred))
print("To‘liq hisobot:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
import matplotlib.pyplot as plt

# Xususiyatlar ahamiyati
importances = model.feature_importances_
feature_names = iris.feature_names

# Vizualizatsiya
plt.figure(figsize=(8, 5))
plt.barh(feature_names, importances)
plt.xlabel("Ahamiyat")
plt.title("Xususiyatlarning ahamiyati (Feature Importance)")
plt.show()
```
